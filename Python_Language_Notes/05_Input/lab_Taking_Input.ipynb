## a = input("Enter your name\n"):

The input() function pauses the program and waits for the user to type something on the keyboard and press Enter.
The string "Enter your name\n" is the prompt shown to the user. The \n creates a new line, so the user types below the prompt.
Crucial Concept: In Python 3, input() always returns a string (str). Even if you type numbers like 123, Python treats it as text ("123").
This value is stored in the variable a.
print("Hello, " + a + "!"):

This uses string concatenation (joining strings together with +).
It combines "Hello, ", the value inside a, and "!" to print a greeting.
type(a):

The type() function returns the data type of the variable.
Since input() always returns a string, this will return <class 'str'>.

## The eval() Function

The eval() function stands for evaluate. It takes a string and executes it as if it were Python code.

````python
# If the user types: 5 + 5
user_input = input("Enter a math problem: ") 
# user_input is the string "5 + 5"

calculated_value = eval(user_input)
# calculated_value is the integer 10

print(calculated_value)

'''Checking Data Type Using type()
Always show details

type(a)
‚úÖ Output
Always show details

<class 'str'>'''
````

````python
b=eval(input("Enter Your age: \n"))
print("Your age is ",b)
type(b) #it will return 
````
