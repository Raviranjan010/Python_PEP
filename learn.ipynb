## Rules for Identifiers in Python
An identifier is the name used to identify a variable, function, class, or object in Python.

- **Must not be a keyword**: Identifiers cannot use Python reserved words such as `if`, `for`, `while`, `class`, etc.
- **Must start with a letter or an underscore (_)**: It cannot start with a number.
- **Can contain only letters, digits, and underscores**: Allowed: a–z, A–Z, 0–9, _
- **Special characters like @, #, $, %, etc. are not allowed.**
- **No spaces are allowed**: Identifiers must be written as a single continuous word.
- **Case-sensitive**: `total`, `Total`, and `TOTAL` are treated as different identifiers.
- **Length is not strictly limited**: Python does not enforce a maximum length, but identifiers longer than 72 characters are discouraged for readability and good coding practice.
